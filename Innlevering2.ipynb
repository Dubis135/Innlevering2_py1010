```python
## Oppgave 5

##Figuren består av halvsirkel og trekant. Ved beregning av omkrets må vi huske å trekke kateten som ligger inne i figuren.

import numpy as np
a=float(input("Skriv inn verdien til a")) ## Diameter i øverste halvsirkel og en av katetene i trekanten
b=float(input("Skriv inn verdien til b")) ## Katet i trekanten
radius=a/2

## Regne ut arealet

areal_sirkel=np.pi * radius**2
areal_trekant= (a*b)/2 ## Areal i rettvinklet trekant
areal_figur= round((areal_sirkel / 2) + areal_trekant,2)

print("Arealet til figuren er", areal_figur, "cm2.")

## Regne ut omkretsen

omkrets_sirkel = 2*np.pi*radius
hypotenose=np.sqrt(a**2 + b**2)
omkrets_figur = round((omkrets_sirkel/2) + b + hypotenose,2)

print("Omkretsen til figuren er", omkrets_figur, "cm.")
```
